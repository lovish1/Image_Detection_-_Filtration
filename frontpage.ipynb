```python
from PIL import ImageTk,Image
import tkinter
from tkinter import Tk,Label
import cv2
global panelA
global panelB
global panelC
def firstpage():
    root=Tk()
    uname=StringVar()
    pswd=StringVar()
    def success():
        root.destroy()
        var=Toplevel()
        def back():
            root.destroy()
            firstpage()
        var.title('Second Page')
        var.config(bg='powderblue',width=800,height=800)
        image=cv2.imread('car.png')
        gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
        edged=cv2.Canny(gray,50,100)
        image=Image.fromarray(image)
        gray=Image.fromarray(gray)
        edged=Image.fromarray(edged)
        image=ImageTk.PhotoImage(image)
        gray=ImageTk.PhotoImage(gray)
        edged=ImageTk.PhotoImage(edged)
        def org():
            panelA=Label(image=image)
            panelA.image=image
            panelA.grid(row=0,column=2)
        def filtered():
            panelB=Label(image=gray)
            panelB.image=gray
            panelB.grid(row=0,column=3)
        bkbtn=Button(var,text='BACK',command=back,bg='Black',fg='white',font=("bold"))
        bn1=Button(var,text='Original Image',command=org,bg='snow3',fg='black',font=("bold"))
        bn1.grid(row=1,column=1)
        bn2=Button(var,text='Filtered Image',command=filtered,bg='snow3',fg='black',font=("bold"))
        bn2.grid(row=2,column=1)
        #bn3=Button(var,text='Edge Detection',bg='snow3',fg='black',font=("bold"))
        #bn3.grid(row=3,column=1)
    def login():
        if(ub.get()=='secttron' and pb.get()=='1234'):
            success()
        else:
            root.destroy()
    def shut():
        root.destroy()
    can=Canvas(root,bg='SkyBlue',width=860,height=480)
    can.pack()
    h1=Label(root,text='Image Filtration & Edge Detection',bg='skyblue',font=("aerial 20 bold italic"))
    h1.pack()
    i1=ImageTk.PhotoImage(file="pic4.png")
    can.create_image(8,490,anchor=SW,image=i1)
    can.create_window(640,30,window=h1)
    usm=Label(root,text='Username:',bg='SkyBlue',borderwidth=3,relief='ridge',font=("Times 18 italic"))
    can.create_window(600,160,window=usm)
    ub=Entry(root,textvariable=uname,borderwidth=3,relief='sunken')
    can.create_window(730,160,window=ub)
    pwd=Label(root,text='Password:',bg='SkyBlue',borderwidth=3,relief='ridge',font=("Times 18 italic"))
    can.create_window(600,230,window=pwd)
    pb=Entry(root,textvariable=pswd,borderwidth=3,relief='sunken')
    can.create_window(730,230,window=pb)
    btn=Button(root,text='LOGIN',command=login,bg='dark slate gray',fg='goldenrod',font=('bold'))
    can.create_window(660,290,window=btn)
    btn2=Button(root,text='QUIT',command=shut,bg='dark slate gray',fg='white',font=('bold'))
    can.create_window(820,460,window=btn2)
    root.mainloop()
firstpage()
```
